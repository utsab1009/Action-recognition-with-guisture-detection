```python
action_images_dict = {
    'bye': list(data_dir.glob('Bye/*')),
    'fine': list(data_dir.glob('Fine/*')),
    'hello': list(data_dir.glob('Hello/*')),
    'no': list(data_dir.glob('No/*')),
    'yes': list(data_dir.glob('Yes/*')),
    'thanks': list(data_dir.glob('Thanks/*'))
}
```

```python
action_labels_dict = {
    'bye': 0,
    'fine': 1,
    'hello': 2,
    'no': 3,
    'yes': 4,
    'thanks':5
}
```

```python
action_images_dict['yes'][:5]
```

```python
X, y = [], []

for action_name, images in action_images_dict.items():
    for image in images:
        img = cv2.imread(str(image))
        resized_img = cv2.resize(img,(180,180))
        X.append(resized_img)
        y.append(action_labels_dict[action_name])
```

```python
import numpy as np
X = np.array(X)
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python

X_train_scaled = X_train / 255
X_test_scaled = X_test / 255
```

```python
num_classes = 6

model = Sequential([
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(128, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(256, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(512, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes)
])

model.compile(optimizer='adam',
              loss=tensorflow.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
              
r=model.fit(X_train_scaled, y_train, epochs=50)
```

```python
p=model.evaluate(X_test_scaled,y_test)
```

```python
plt.plot(r.history['loss'], label=' loss')
plt.plot(r.history['accuracy'], label=' acc')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Rate')
```
